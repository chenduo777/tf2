```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import initializers
from preprocessing import parse_aug_fn, parse_fn
```

```python
summary_writer = tf.summary.create_file_writer('lab7-logs-summary')
```

```python
x = np.linspace(0,2*np.pi,100)
data = np.sin(x)
with summary_writer.as_default():
    for i,y in enumerate(data):
        tf.summary.scalar("Sinc Function", y,step=i)
```

```python
def read_png_image(img):
    img = tf.io.read_file(img)
    img = tf.image.decode_png(img, channels=1)
    return img

img = read_png_image('360_F_173017473_xRG5TrDyf7elY5PPC7Xt89Ar3imtHwF4.png')
plt.imshow(img)
```

```python
with summary_writer.as_default():
    tf.summary.image("plane", [img], step=0)
```

```python
texts ={
    "name": "John Doe",
    "ID": 123456,
    "image": "ssda.png"
}

with summary_writer.as_default():
    for i,text in enumerate(texts):
        tf.summary.text("text", str(text), step=i)
```
